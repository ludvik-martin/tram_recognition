```python
DATA_PATH = '../data'
```

```python
def load_samples_from_subdirs(main_dir):
    sample_files = {tram_type:'{}/{}/{}'.format(main_dir, tram_type, random.choice(os.listdir(f'{main_dir}/{tram_type}'))) 
               for tram_type in sorted(os.listdir(main_dir))}
    samples = {tram_type: librosa.load(sample_files[tram_type], mono=True)[0] for tram_type in sample_files.keys()}
    return samples, list(sample_files.values())

def load_samples(main_dir):
    sample_files = [f'{main_dir}/{sample}' for sample in random.sample(os.listdir(main_dir), 5)]
    samples = [librosa.load(sample, mono=True)[0] for sample in sample_files] 
    return samples, sample_files
```

```python
# to have repeatable values
random.seed(333)

acc_samples, acc_files = load_samples_from_subdirs('{}/{}'.format(DATA_PATH, 'acc'))
break_samples, break_files = load_samples_from_subdirs('{}/{}'.format(DATA_PATH, 'break'))
neg_samples, neg_files = load_samples(f'{DATA_PATH}/neg')
```

Let's prepare data.

```python
def normalized_spectrogram_img(wave):
    # cut the end of the wave
    wave = wave[:50000]
    spectrogram = librosa.stft(wave,
                   n_fft      = 1024)
    power_spectrogram = np.abs(spectrogram)**2
    spectrogram_db = librosa.amplitude_to_db(power_spectrogram, ref=np.max)
    # keep only interesting frequencies
    spectrogram_db_band = spectrogram_db[:100] 
    return spectrogram_db_band

def read_normalized_spectrogram_img(file):
    wave = librosa.load(file, mono=True)[0]
    return normalized_spectrogram_img(wave)
```

```python
le = preprocessing.LabelEncoder()
le2 = preprocessing.LabelEncoder()
def prepare_data(main_dir):
    X = []
    y_tram_category = []
    y_acc = []
    
    for acc in ['acc', 'break']:
        for category in os.listdir(f'{main_dir}/{acc}'):
            for file in os.listdir(f'{main_dir}/{acc}/{category}'):
                full_path = f'{main_dir}/{acc}/{category}/{file}'
                X.append(read_normalized_spectrogram_img(full_path))
                y_tram_category.append(category)
                y_acc.append(acc)
    
    for file in os.listdir(f'{main_dir}/neg'):
        full_path = f'{main_dir}/neg/{file}'
        X.append(read_normalized_spectrogram_img(full_path))
        y_tram_category.append('neg')
        y_acc.append('neg')
                
    # normalize
    X = np.array(X)
    X = np.expand_dims(X, axis=-1)
    X_normalized = (X - X.mean()) / (X.max() - X.min())
    # the categories have to be numbers, let's preprocess them
    le.fit(y_tram_category)
    le2.fit(y_acc)
    y_tram_category = le.transform(y_tram_category)
    y_acc = le2.transform(y_acc)
    return np.array(X_normalized), np.array(y_tram_category), np.array(y_acc)
```

```python
X, y_tram_category, y_acc = prepare_data(DATA_PATH)
X_train, X_test, y_tram_category_train, y_tram_category_test, y_acc_train, y_acc_test = \
    train_test_split(X, y_tram_category, y_acc, test_size=0.2)
```

```python
def get_batch(data, labels, batch_size):
    index = np.random.randint(0, np.shape(data)[0], batch_size)
    return data[index, :], labels[index]
```

```python
dropout_rate = 0.5

class CNN(tf.keras.Model):
    def __init__(self):
        super().__init__()
        self.conv1 = tf.keras.layers.Conv2D(
            filters=32,
            kernel_size=[5, 5],
            padding="same",
            activation=tf.nn.relu
        )
        self.pool1 = tf.keras.layers.MaxPool2D(pool_size=[2, 2], strides=2)
        self.drop1 = tf.keras.layers.Dropout(dropout_rate)
        self.conv2 = tf.keras.layers.Conv2D(
            filters=64,
            kernel_size=[3, 3],
            padding="same",
            activation=tf.nn.relu
        )
        self.pool2 = tf.keras.layers.MaxPool2D(pool_size=[2, 2], strides=2)
        self.conv3 = tf.keras.layers.Conv2D(
            filters=64,
            kernel_size=[3, 3],
            padding="same",
            activation=tf.nn.relu
        )
        self.drop2 = tf.keras.layers.Dropout(dropout_rate)
        self.pool3 = tf.keras.layers.MaxPool2D(pool_size=[2, 2], strides=2)
        self.flatten = tf.keras.layers.Flatten()
        self.dense1 = tf.keras.layers.Dense(units=1024, activation=tf.nn.relu)
        # 4 types of tram + neg
        self.dense2 = tf.keras.layers.Dense(units=5)

    def call(self, inputs, training):
        x = self.conv1(inputs)                  
        x = self.pool1(x)
        x = self.drop1(x, training=training)
        x = self.conv2(x)
        x = self.pool2(x)
        x = self.drop2(x, training=training)
        x = self.conv3(x)
        x = self.pool3(x)
        x = self.flatten(x)
        x = self.dense1(x)
        x = self.dense2(x)
        return x

    def predict(self, inputs):
        logits = self(inputs, training=False)
        return tf.argmax(logits, axis=-1)
```

```python
learning_rate = 1e-3
batch_size = 32
num_batches = 200

model = CNN()
optimizer = tf.compat.v1.train.AdamOptimizer(learning_rate=learning_rate)
```

```python
for batch_index in range(num_batches):
    X, y = get_batch(X_train, y_tram_category_train, batch_size=batch_size)
    with tf.GradientTape() as tape:
        y_logit_pred = model(tf.convert_to_tensor(X), training=True)
        loss = tf.reduce_mean(tf.losses.sparse_categorical_crossentropy(y_true=y, y_pred=y_logit_pred, from_logits=True))
        if (batch_index % 10 == 0):
            print("batch %d: loss %f" % (batch_index, loss.numpy()))
    grads = tape.gradient(loss, model.variables)
    optimizer.apply_gradients(grads_and_vars=zip(grads, model.variables))
```

```python
y_pred = model.predict(X_test).numpy()
print("test accuracy: %f" % (sum(y_pred == y_tram_category_test) / len(y_pred)))
```

However, we did not try to recognize if the tram if accelerating or breaking. Let's do that. To simplify training, le'ts try to do multi-task training, with the two tasks being classification of the tram type and classification of the acc/break/neg. Both tasks would share the convolutional features, which might improve the performance of the both tasks.

```python
def get_batch_multitask(data, labels_tram_cat, labels_acc, batch_size):
    index = np.random.randint(0, np.shape(data)[0], batch_size)
    return data[index, :], labels_tram_cat[index], labels_acc[index]
```

```python
dropout_rate = 0.5

class CNN2(tf.keras.Model):
    def __init__(self):
        super().__init__()
        self.conv1 = tf.keras.layers.Conv2D(
            filters=32,             # Numbers of convolution kernels.
            kernel_size=[5, 5],     # Size of the receptive field.
            padding="same",         # Padding strategy.
            activation=tf.nn.relu   # Activation function.
        )
        self.pool1 = tf.keras.layers.MaxPool2D(pool_size=[2, 2], strides=2)
        self.drop1 = tf.keras.layers.Dropout(dropout_rate)
        self.conv2 = tf.keras.layers.Conv2D(
            filters=64,
            kernel_size=[3, 3],
            padding="same",
            activation=tf.nn.relu
        )
        self.pool2 = tf.keras.layers.MaxPool2D(pool_size=[2, 2], strides=2)
        self.conv3 = tf.keras.layers.Conv2D(
            filters=64,
            kernel_size=[3, 3],
            padding="same",
            activation=tf.nn.relu
        )
        self.drop2 = tf.keras.layers.Dropout(dropout_rate)
        self.pool3 = tf.keras.layers.MaxPool2D(pool_size=[2, 2], strides=2)
        self.flatten = tf.keras.layers.Flatten()
        self.dense1 = tf.keras.layers.Dense(units=1024, activation=tf.nn.relu)
        # 4 types of tram + neg
        self.dense2 = tf.keras.layers.Dense(units=5)
        self.dense3 = tf.keras.layers.Dense(units=1024, activation=tf.nn.relu)
        # acc/break/neg
        self.dense4 = tf.keras.layers.Dense(units=3)

    def call(self, inputs, training):
        x = self.conv1(inputs)                  
        x = self.pool1(x)
        x = self.drop1(x, training=training)
        x = self.conv2(x)
        x = self.pool2(x)
        x = self.drop2(x, training=training)
        x = self.conv3(x)
        x = self.pool3(x)
        x = self.flatten(x)
        x1 = self.dense1(x)
        x1 = self.dense2(x)
        x2 = self.dense3(x)
        x2 = self.dense4(x)
        # logits for tram classes, logits for acc/break/neg
        return x1, x2

    def predict(self, inputs):
        logits_tram_cat, logits_acc = self(inputs, training=False)
        return tf.argmax(logits_tram_cat, axis=-1), tf.argmax(logits_acc, axis=-1)
```

```python
learning_rate = 1e-3
batch_size = 32
num_batches = 200

model = CNN2()
optimizer = tf.compat.v1.train.AdamOptimizer(learning_rate=learning_rate)
```

```python
for batch_index in range(num_batches):
    X, y_tram_cat, y_acc = get_batch_multitask(X_train, y_tram_category_train, y_acc_train, batch_size=batch_size)
    with tf.GradientTape() as tape:
        y_logit_tram_cat, y_logit_acc = model(tf.convert_to_tensor(X), training=True)
        
        loss_tram_cat = tf.reduce_mean(tf.losses.sparse_categorical_crossentropy(y_true=y_tram_cat, 
                    y_pred=y_logit_tram_cat, from_logits=True))
        loss_acc = tf.reduce_mean(tf.losses.sparse_categorical_crossentropy(y_true=y_acc, 
                    y_pred=y_logit_acc, from_logits=True))
        loss = loss_tram_cat + loss_acc
        if (batch_index % 10 == 0):
            print("batch %d: loss %f" % (batch_index, loss.numpy()))
    grads = tape.gradient(loss, model.variables)
    optimizer.apply_gradients(grads_and_vars=zip(grads, model.variables))
```

```python
y_tram_cat_pred, y_acc_pred = model.predict(X_test)
print("test accuracy for tram category: %f" % (sum(y_tram_cat_pred.numpy() == y_tram_category_test) / len(y_tram_category_test)))
print("test accuracy for acc/break/neg: %f" % (sum(y_acc_pred.numpy() == y_acc_test) / len(y_acc_test)))
```

The results show that multi-task learning with combined loss function actually helped also with the tram category recognition.

Future improvements for CNN approach:  
- The number of samples for each category should be similar and at leat 1000 (rule of thumb)
- We designed very simple CNN. I would consider using convolutional layers from out-of-the-box deep CNN (Inception-ResNet, VGG)  pretraiend on much bigger dataset (e.g. COCO, Imagenet) and use transfer learning to fine-tune it on our tram dataset  
- I would recommend to take samples from different distances and by different microphones, to make the model more robust 
- Note: Usual data augmentation approaches cannot be used, e.g. shifting or mirroring the spectrogram would mean very different sample
